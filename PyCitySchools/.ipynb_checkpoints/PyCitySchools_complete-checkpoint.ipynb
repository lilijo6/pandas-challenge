```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Performing the calculations to find the district summary from the complete dataframe.
school_count = len(school_data_complete['school_name'].unique())

total_students = len(school_data_complete['student_name'])

total_budget = sum(school_data_complete['budget'].unique())

average_math_score = school_data_complete['math_score'].mean()

average_reading_score = school_data_complete['reading_score'].mean()

passingmath = len(school_data_complete.loc[school_data_complete['math_score'] >= 70]) / total_students 

passingreading = len(school_data_complete.loc[school_data_complete['reading_score'] >= 70]) / total_students 

passingboth = len(school_data_complete.loc[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]) / total_students 
```

```python

DistrictSummary_df = pd.DataFrame({"Total Schools" : [school_count],
                                  "Total Students" : [total_students],
                                  "Total Budget USD" : [total_budget],
                                  "Average Math Score" : average_math_score,
                                  "Average Reading Score" : average_reading_score,
                                  "% Passing Math" : passingmath,
                                  "% Passing Reading": passingreading,
                                  "% Passing Both": passingboth})

DistrictSummary_df["Total Budget USD"]=DistrictSummary_df["Total Budget USD"].map("${:,.2f}".format)
DistrictSummary_df["% Passing Math"]=DistrictSummary_df["% Passing Math"].map("{:,.2%}".format)
DistrictSummary_df["% Passing Reading"]=DistrictSummary_df["% Passing Reading"].map("{:,.2%}".format)
DistrictSummary_df["% Passing Both"]=DistrictSummary_df["% Passing Both"].map("{:,.2%}".format)
DistrictSummary_df
```

```python
#Groouping by school and getting the average for the reading score, math score and getting the budget for each school
schools_grouped = average_schools_df.groupby(["school_name"])
average_schools_grouped = schools_grouped.mean()
average_schools_grouped
```

```python
#Creating a series for each column 
Average_budget = average_schools_grouped["budget"]
Average_reading = average_schools_grouped["reading_score"]
Average_math = average_schools_grouped["math_score"]
```

```python
#Alternative way of calculating the student count (for comparison only)
student_df = school_data_complete[["school_name","student_name"]]
student_df
```

```python
#Alternative way of calculating the student count (for comparison only)
student_grouped = student_df.groupby(["school_name"])
small_schools_student_df = student_grouped.count()["student_name"]
small_schools_student_df
```

```python
# Calculating per student budget
budget_ps_df = average_schools_grouped["budget"] / small_schools_student_df
budget_ps_df
```

```python
#Creating a new dataframe for the scores only. 
Passingscores_df = school_data_complete[["school_name", "reading_score", "math_score"]]
Passingscores_df.head()
```

```python
#Creating a df for only passing students and grouping by school to calculate the percent passing math
passingmath_df = Passingscores_df[Passingscores_df["math_score"] >= 70]
passingmath_df_grouped = passingmath_df.groupby(["school_name"])
total_math_passed = passingmath_df_grouped["math_score"].count()
percent_passingmath = total_math_passed / small_schools_student_df
percent_passingmath 
```

```python
#Creating a df for only passing students and grouping by school to calculate the percent passing reading
passingreading_df = Passingscores_df[Passingscores_df["reading_score"] >= 70]
passingreading_df_grouped = passingreading_df.groupby(["school_name"])
total_reading_passed = passingreading_df_grouped["reading_score"].count()
percent_passingreading = total_reading_passed / small_schools_student_df
percent_passingreading 
```

```python
#Creating a df for only passing students and grouping by school to calculate the percent passing both
passingboth_df = Passingscores_df[(Passingscores_df["reading_score"] >= 70) & (Passingscores_df["math_score"] >= 70)]
passingboth_df_grouped = passingboth_df.groupby(["school_name"])
total_both_passed = passingboth_df_grouped["math_score"].count()
percent_passingboth = total_both_passed / small_schools_student_df
percent_passingboth 
```

```python
#Creating a dataframe to contain the complete summary for each school
summary_schools_df = pd.DataFrame({"School type": schooltypes, 
                                   "Total Students" : studentcounts,
                                   "Total School Budget": Average_budget,
                                   "Per Student Budget" : budget_ps_df,
                                   "Average Math Score" : Average_math,
                                   "Average Reading Score" : Average_reading,
                                   "% Passing Math" : percent_passingmath,
                                   "% Passing Reading" : percent_passingreading ,
                                   "% Overall Passing" : percent_passingboth})
summary_schools_df
```

```python
#Formatting the percentage values and the currency values
summary_schools_df["Total School Budget"]=summary_schools_df["Total School Budget"].map("${:,.2f}".format)
summary_schools_df["Per Student Budget"]=summary_schools_df["Per Student Budget"].map("${:,.2f}".format)
summary_schools_df["% Passing Math"]=summary_schools_df["% Passing Math"].map("{:,.2%}".format)
summary_schools_df["% Passing Reading"]=summary_schools_df["% Passing Reading"].map("{:,.2%}".format)
summary_schools_df["% Overall Passing"]=summary_schools_df["% Overall Passing"].map("{:,.2%}".format)
summary_schools_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Reseting the Name of the school as the index
summary_schools_df = summary_schools_df.reset_index()
summary_schools_df.head()
```

```python
#Creating a new dataframe that contains only the relevant variables
school_spending = summary_schools_df[["Per Student Budget", "Average Math Score", "Average Reading Score", "% Passing Math", \
                                   "% Passing Reading", "% Overall Passing"]]
school_spending.head()
```

```python
# Converting the "Per Student Budget" and all percent column to floats
school_spending["Per Student Budget"] = school_spending["Per Student Budget"].replace(
    '\$', '', regex=True).astype('float')
school_spending["% Passing Math"] = school_spending["% Passing Math"].replace(
    '%', '', regex=True).astype('float')
school_spending["% Passing Reading"] = school_spending["% Passing Reading"].replace(
    '%', '', regex=True).astype('float')
school_spending["% Overall Passing"] = school_spending["% Overall Passing"].replace(
    '%', '', regex=True).astype('float')
```

```python
#Checking the new data
school_spending.head()
```

```python
#Verifying the types of objects in the dataframe
school_spending.dtypes
```

```python
#Figuring out the min and max values of the Per Student Budget
print(school_spending["Per Student Budget"].max())
print(school_spending["Per Student Budget"].min())
```

```python
#Calculating a range
rangespending = school_spending["Per Student Budget"].max() - school_spending["Per Student Budget"].min()
rangespending 
```

```python
#Creating the bins to contain the values
bins = [0, 585, 630, 645, 676]

group_labels = [" <$584", "$585-629", "$630-644", "$645-675"]
```

```python
#Slice the data and place it to bins
pd.cut(school_spending["Per Student Budget"], bins, labels=group_labels).head()
```

```python
# Replacing the data in the Per Student budget with the bins created
school_spending["Per Student Budget"] = pd.cut(school_spending["Per Student Budget"], bins, labels=group_labels)
school_spending.head()
```

```python
#Renaming the Per Student budget column to match the bin info
school_spending = school_spending.rename(columns={"Per Student Budget":"Spending Ranges (Per Student)"})
school_spending.head()
```

```python
#Grouping by bins to create new summary table
spending_grouped = school_spending.groupby("Spending Ranges (Per Student)")
spending_result = spending_grouped.mean()

#Formatting
spending_result["Average Math Score"]=spending_result["Average Math Score"].map("{:,.2f}".format)
spending_result["Average Reading Score"]=spending_result["Average Reading Score"].map("{:,.2f}".format)
spending_result["% Passing Math"]=spending_result["% Passing Math"].map("{:,.2f}".format)
spending_result["% Passing Reading"]=spending_result["% Passing Reading"].map("{:,.2f}".format)
spending_result["% Overall Passing"]=spending_result["% Overall Passing"].map("{:,.2f}".format)

spending_result
```
